Data (all_ticker_2015_25.pkl) is a list of pd series. Each dictionary contains data for a specific ticker symbol, with keys like 'Open', 'High', 'Low', 'Close', and 'Volume'. The goal is to convert this list of dictionaries into a list of pandas DataFrames, where each DataFrame corresponds to a ticker and contains the relevant data.

### MultiIndex for columns
Top level: Open, High, Low, Close, Volume
Second level: ticker symbols (A, AA, AAGR, etc.)

`data["Open"]` with return a sub dataframe where the columns are all the tickers, and then rows are all the open prices.
